### 2.2.2 Positional Encoding

```python
import japanize_matplotlib
import matplotlib.pyplot as plt
import numpy as np

K = 50  # Maximum length of word sequence
D = 64  # Dimension of embeddings

# Initialize the positional encoding matrix
pos_enc = np.empty((K, D))

for i in range(K):  # Loop over word position i
    for k in range(D // 2):  # Loop over values of k
        theta = i / (10000 ** (2 * k / D))
        pos_enc[i, 2 * k] = np.sin(theta)
        pos_enc[i, 2 * k + 1] = np.cos(theta)
```

```python
# Display the matrix as an image
im = plt.imshow(pos_enc)
plt.xlabel("次元 (Dimension)")  # Set label for the X-axis
plt.ylabel("位置 (Position)")  # Set label for the Y-axis
plt.colorbar(im)  # Add a bar to indicate the correspondence between values and colours
plt.show()
```

```python
# Calculate the dot product between positional encodings
dot_matrix = np.matmul(pos_enc, pos_enc.T)
# Display the matrix as an image
im = plt.imshow(dot_matrix, origin="lower")
plt.xlabel("位置 (Position)")
plt.ylabel("位置 (Position)")
plt.colorbar(im)
plt.show()
```
